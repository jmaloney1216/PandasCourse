```python
import pandas as pd
import numpy as np
```

# Changing Options with Attributes and Dot Syntax

```python
data = np.random.randint(0,100, [1000, 50])
```

```python
df = pd.DataFrame(data)
```

```python
df.tail(2)
```

```python
pd.options.display.max_rows
```

```python
pd.options.display.max_rows = 18
```

```python
df
```

```python
pd.options.display.max_columns = 4
```

```python
df
```
