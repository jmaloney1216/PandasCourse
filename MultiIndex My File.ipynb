# Extract Rows from a MultiIndex DataFrame

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates = ['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True) #This will greatly enhance the speed of processing
bigmac.head(3)
```

```python
bigmac.loc[('2010-01-01', 'Brazil'), 'Price in US Dollars'] #Must pass a tuple
```

```python
bigmac.loc[('2015-07-01', 'Chile'), 'Price in US Dollars']
```

```python
bigmac.ix[('2016-01-01', 'China'), 'Price in US Dollars']
```

```python
bigmac.ix[('2016-01-01', 'China'), 0]
```

# The .transpose() Method and MultiIndex on Column Level

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates = ['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True) #This will greatly enhance the speed of processing
bigmac.head(3)
```

```python
bigmac = bigmac.transpose() #swaps the axes
bigmac.head()
```

```python
bigmac.loc['Price in US Dollars']
```

```python
bigmac.loc['Price in US Dollars', ('2016-01-01', 'Denmark')]
```

# The .swaplevel() Method

```python
bigmac = pd.read_csv('bigmac.csv', parse_dates = ['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True) #This will greatly enhance the speed of processing
bigmac.head(3)
```

```python
#swaps levels in a multiindex
bigmac = bigmac.swaplevel()
bigmac.head(3)
```

# The .pivot() Method

```python
sales = pd.read_csv('salesmen.csv', parse_dates = ['Date'])
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head(3)
```

```python
sales['Salesman'].value_counts()
```

```python
sales.pivot(index = 'Date', columns = 'Salesman', values = 'Revenue')
```

# The .pivot_table() Method

```python
foods = pd.read_csv('foods.csv')
foods.head(3)
```

```python
foods.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'mean')
```

```python
foods.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'sum')
```

```python
foods.pivot_table(values = 'Spend', index = 'Item', aggfunc = 'mean')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], aggfunc = 'mean')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], columns = 'City', aggfunc = 'mean')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], columns = ['Frequency', 'City'], aggfunc = 'mean')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], columns = 'City', aggfunc = 'max')
```

```python
pd.pivot_table(data = foods, values = 'Spend', index = ['Gender', 'Item'], columns = 'City', aggfunc = 'max')
```
